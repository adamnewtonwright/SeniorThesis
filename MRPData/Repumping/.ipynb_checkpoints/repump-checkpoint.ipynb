```python
#mag field current = .12 A
repump = 0.00
master = 1.53

angles = np.array([0,90])
dv =     np.array([71.2,61.6])

plt.figure('0repump')
plt.plot(angles,dv,label = '0% Repumper Continuous')
plt.grid(True)
plt.xlabel('Angle')
#plt.title('0% Repumper Repump')
plt.legend()

plt.figure('0repump scaled')
plt.plot(angles,dv/dv[0],'.--',markersize = 10.0, label = '0% Repumper Continuous')
plt.grid(True)
plt.xlabel('Angle (degrees)')
plt.ylabel('Fluorescence as Percent of Max')
#plt.title('0% Repumper', fontsize = 18)
plt.legend()
```

# High power on continuous

```python
#mag field current = .12 A
#Continuous
repump = 1.00
master = 6.00

angles = np.array([0,90])
dv =     np.array([166,142])
dv = dv/master

dvrepump = np.array([192,172])
dvrepump = dvrepump / (master)

plt.figure('0repump')
plt.plot(angles,dv,'.--', label = '0% Repumper Continuous')
plt.plot(angles,dvrepump,'.--', label = '20% Repumper Continuous')
plt.grid(True)
plt.xlabel('Angle')
#plt.title('0% Repump')
plt.ylabel('Fluorescence / power (dv/W)')
plt.legend()

plt.figure('0repump scaled')
plt.plot(angles,dv/dv[0], label = '0% Repumper Continuous')
plt.grid(True)
plt.xlabel('Angle')
plt.ylabel('Fluorescence / power (dv/W)')
#plt.title('0% Repump Scaled')
plt.legend()
```

```python
#mag field current = .12 A
#pulsed
repump = 0.00
master = 2.40

angles = np.array([0,90])
dv =     np.array([74.4,69.6])
dv = dv/master

plt.figure('0repump')
plt.plot(angles,dv,'.--', label = '0% Repumper Pulsed')
plt.grid(True)
plt.xlabel('Angle (degrees)')
plt.legend()
plt.savefig('transat.pdf')

plt.figure('0repump scaled')
plt.plot(angles,dv/dv[0], label = '0% Repumper Pulsed')
plt.grid(True)
plt.xlabel('Angle (degrees)')
plt.legend()
```
