```python
import numpy as np
import matplotlib.pyplot as plt
```

### Procedure:

For each duty cycle in {20%,40%,60%,80%}, I will measure the fluorescence from the main laser pulsed at maximum power, then measure the fluorescence from the main laser CW at maximum power, and then reduce the power of the main laser CW to the power of the pulsed and measure fluorescence. All fluorescence will be divided by power before plotted.

This is all without a repumper

## 20% Duty Cycle

```python
# Continuous .5 mW

angles = np.array([0,90])
plt.figure('20')
power = .49 #mW
dv = np.array([25.6,22])
eff = dv/power
plt.plot(angles,eff, '.--', label =  'Cont; P = 0.5 mW')

plt.figure('20scaled')
plt.plot(angles,eff/eff[0], '.--', label = 'Cont; P = 0.5 mW')
```

```python
# Continuous 3.08 mW

plt.figure('20')
power = 3.08 #mW
dv = np.array([76.8,65.6])
eff = dv/power
plt.plot(angles,eff, '.--', label = 'Cont; P = 3.08 mW')

plt.figure('20scaled')
plt.plot(angles,eff/eff[0], '.--', label = 'Cont; P = 3.08 mW')
```

```python
# 20% Pulsed .55 mW

plt.figure('20')
power = 0.55 #mW
dv = np.array([22.4,22])
eff = dv/power
plt.plot(angles,eff, '.--', label = '20% Pulsed; P = 0.55 mW')

plt.figure('20scaled')
plt.plot(angles,eff/eff[0], '.--', label = '20% Pulsed; P = 0.55 mW')
```

```python
#Figure 

plt.figure('20')
plt.legend()
plt.grid(True)
plt.xlabel('Angle (degrees)')
plt.ylabel('Return Efficiency (dv/mW)')

plt.figure('20scaled')
plt.legend()
plt.grid(True)
plt.xlabel('Angle (degrees)')
plt.ylabel('Return Efficiency as Percent of Max (dv/mW)')
plt.show()
```
